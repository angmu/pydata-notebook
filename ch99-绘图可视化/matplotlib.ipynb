### matplotlib绘图

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 等差数列
x = np.linspace(0, 10, 100)
```

```python
siny = np.sin(x)
```

```python
plt.plot(x,siny)
# plt.show()
```

```python
cosy = np.cos(x)
```

```python
plt.plot(x, siny)
plt.plot(x, cosy, color='red', linestyle='--')
# plt.xlim(-2, 12)
# 调节x,y的范围
plt.axis([-1,11,-2,2])
plt.xlabel('x轴')
plt.ylabel('y value')
```

```python
plt.plot(x, siny, label='sin(x)')
plt.plot(x, cosy, color='red', linestyle='--', label='cos(x)')
plt.legend() # 添加标注
# 调节x,y的范围
plt.xlabel('x轴')
plt.ylabel('y value')
# plt.title('welcome to the ML world')
```
