## 披萨店顾客数据

```python
import seaborn as sns
```

```python
tips=sns.load_dataset('tips')
```

## 1.带状图

离散数据 和 连续数据 之间的关系

```python
sns.stripplot(data=tips, x='day', y='total_bill', jitter=False)
```

```python
sns.stripplot(data=tips, x='day', y='total_bill', jitter=True) #jitter 抖动
```

## 2.蜂群图

### 离散数据 和 连续数据 之间的关系-密度排列

带状图的补充

不允许在一起

```python
sns.swarmplot(data=tips, x='day', y='total_bill')
```

```python
tips.head()
```

## 3.Hue-分组参数

分析每天中，午餐和晚餐的账单分布

```python
sns.swarmplot(data=tips, x='day', y='total_bill', hue='time')
```

每天的付账人群中的性别分布

```python
sns.swarmplot(data=tips, x='day', y='total_bill', hue='sex')
```

```python
sns.swarmplot(data=tips, x='day', y='total_bill', hue='size')
```

## 提琴图
