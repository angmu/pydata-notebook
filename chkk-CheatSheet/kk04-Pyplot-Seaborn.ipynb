### 结构化数据绘制散点图

原始2个特征，之间的比较

x相同，y上下变动

```python
data={
    'a':np.arange(50),
    'c':np.random.randint(0,50,50),
    'd':np.random.rand(50)
}
```

## plt.scatter()

```python
plt.scatter('a', 'd', data=data)
plt.xlabel('a数据')
plt.ylabel('b数据')
```

```python
data['b']=np.abs(data['d'])
```

```python
data
```

```python
plt.scatter('a', 'd', data=data, c='c', marker='>') # 散点图的样式
```

```python
plt.scatter('c', 'd', data=data, c='c', marker='>') # 散点图的样式, c --- colorful五颜六色随值变化
```

## 柱状图绘制

```python
names=['A类型', 'B类型', 'C类型']
values=[1,10,100]
```

```python
# plt.bar([0, 1, 2], values)
plt.bar(range(len(names)), values)
plt.xticks(range(len(names)), names)
plt.title('离散数据散点图')
```

## 子图-SubPlot
1. 将一个画布进行切分（Figure）
2. 将切分后的图分配到固定的位置
3. 将图可以设置成固定的大小


行数，列数，位置

栅格系统

```python
# plt.figure()
plt.subplot(131)
plt.bar(names, values,color='r')
plt.subplot(235)
plt.scatter(names, values,color='y')
plt.subplot(233)
plt.plot(names, values,color='g')
plt.title('离散数据的柱状图、散点图、折线图')
```
